```python
import pandas as pd
import os
import certifi
os.environ['SSL_CERT_FILE'] = certifi.where()
df=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df=df[0]
df
```

```python
df=df.set_index('Symbol')
df
```

## Horizonatal addition (merge)

```python
df1=df[['Security','GICS Sector','GICS Sub-Industry']]
df1
```

```python
df2=df.iloc[:,-3:]
df2
```

```python
pd.merge(df1,df2,on='Symbol')
```

```python
import numpy as np
np1=np.random.randint(100,200,25).reshape(5,5)
data1=pd.DataFrame(np1,index=['r1','r2','r3','r4','r5'],columns=['c1','c2','c3','c4','c5'])
data1
```

```python
import numpy as np
np1=np.random.randint(1000,2000,25).reshape(5,5)
data2=pd.DataFrame(np1,index=['r1','r2','r3','r4','r6'],columns=['t1','t2','t3','t4','t5'])
data2
```

```python
data1.index.name='i1'
data2.index.name='i1'
pd.merge(data1,data2,on='i1',how='outer')
```
