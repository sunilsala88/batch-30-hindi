```python
import pandas as pd
import os
import certifi
os.environ['SSL_CERT_FILE'] = certifi.where()
df=pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')
df=df[2]
df
```

```python
pe={}
for ticker in df['Symbol']:
    stock = finvizfinance(ticker)
    d=stock.ticker_fundament()
    v=d.get('P/E')
    pe.update({ticker:v})
pe
```
